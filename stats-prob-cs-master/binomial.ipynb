```python
import matplotlib.pyplot as plt

# Number of heads
k = [0, 1, 2, 3, 4]

# Corresponding probabilities
prob = [0.0625, 0.25, 0.375, 0.25, 0.0625]

plt.bar(k, prob, color='skyblue')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Probability Distribution of Number of Heads in 4 Coin Tosses')
plt.xticks(k)
plt.show()
```
